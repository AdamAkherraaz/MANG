## **Dataset description**

**Context**<br>
We downloaded on Kaggle a dataset containing MAANG's historical stock market data from 2001 to 2023. We had 5 tables which we reshaped and formatted into a single one using SQL. 

**Dataset content**<br>
This dataset contains 27435 rows and includes the following features: 
 - Open: Price at opening time.
 - Close: Price at closing time.
 - High: Highest value of that stock on that day.
 - Low: Lowest value of that stock on that day.
 - Adj Close : It helps in getting an idea of the fair value of a stock.
 - Volume: Measure of the number of stocks traded over a specified period.
 - Company: Amazon, Apple, Google, Microsoft and Netflix.
 - Month
 - Year
 
Download dataset: https://www.kaggle.com/datasets/soumendraprasad/stock

## **Importing the data**

```python
import pandas as pd
```

```python
data = pd.read_csv('MAANG.csv')
data = data.rename(columns={'c1': 'Open', 'c2': 'Close', 'c3': 'High', 'c4': 'Low', 'c5': 'Adj close', 'c6': 'Volume'})
data.head()
```

## **Visualizing the data**

```python
import matplotlib.pyplot as plt
```

```python
pivot_table = pd.pivot_table(data, values='Open', index=['year'], columns=['company'])
pivot_table.plot(figsize=(10, 6))

plt.title('Evolution of Stock Market Opening Prices by Company')
plt.xlabel('Date')
plt.ylabel('Opening Price')

plt.legend(title='Company', loc='upper left')

plt.show()
```

```python
pivot_table = pd.pivot_table(data, values='Close', index=['year'], columns=['company'])
pivot_table.plot(figsize=(10, 6))

plt.title('Evolution of Stock Market Closing Prices by Company')
plt.xlabel('Date')
plt.ylabel('Closing Price')

plt.legend(title='Company', loc='upper left')

plt.show()
```

```python
pivot_table = pd.pivot_table(data, values='High', index=['year'], columns=['company'])
pivot_table.plot(figsize=(10, 6))

plt.title('Evolution of Stock Market High Prices by Company')
plt.xlabel('Date')
plt.ylabel('High Price')

plt.legend(title='Company', loc='upper left')

plt.show()
```

```python
pivot_table = pd.pivot_table(data, values='Low', index=['year'], columns=['company'])
pivot_table.plot(figsize=(10, 6))

plt.title('Evolution of Stock Market Low Prices by Company')
plt.xlabel('Date')
plt.ylabel('Low Price')

plt.legend(title='Company', loc='upper left')

plt.show()
```

```python
pivot_table2 = pd.pivot_table(data, values='Volume', index=['year'], columns=['company'])
pivot_table2.plot(figsize=(10, 6))

plt.title('Evolution of Volume of Stock Traded by Company')
plt.xlabel('Date')
plt.ylabel('Volume')

plt.legend(title='Company', loc='upper left')

plt.show()
```

```python
data.corr()
```

We can see that the graphs showing the evolution of opening, closing, high and low price look the same. This can be explained by the fact that these values are highly correlated. In fact, if two values are highly correlated, the trends and patterns observed in a graph for one variable will be very similar to the trends and patterns observed in a graph for the other variable.

```python
df_2022 = data[data['year'] == 2021]

pivot_table = pd.pivot_table(df_2022, values='Open', index=['month'], columns=['company'])
pivot_table.plot(figsize=(10, 6))

plt.title('Evolution of Opening Prices by Company for the year 2022')
plt.xlabel('Month')
plt.ylabel('Opening Price')

plt.legend(title='Company', loc='upper right')

plt.show()
```

```python
pivot_table = pd.pivot_table(df_2022, values='Volume', index=['month'], columns=['company'])
pivot_table.plot(figsize=(10, 6))

plt.title('Evolution of Volume of Stocks Traded by Company for the year 2022')
plt.xlabel('Month')
plt.ylabel('Volume')

plt.legend(title='Company', loc='upper right')

plt.show()
```

```python
df_pivot = data.pivot_table(values='Volume', index=['year'], columns='company')

fig, ax = plt.subplots(figsize=(10, 6))
df_pivot.plot.area(stacked=True, ax=ax)

plt.title('Volume of Stocks Traded Over Time')
plt.xlabel('Year')
plt.ylabel('Volume')

plt.show()
```

```python
fig, ax = plt.subplots(figsize=(10, 6))
for company in data['company'].unique():
    plt.scatter(data[data['company'] == company]['Volume'], data[data['company'] == company]['Adj close'], label=company)

ax.set_title('Relationship between Stock Prices and Volume of Stocks Traded')
ax.set_xlabel('Volume')
ax.set_ylabel('Adjusted Close Price')

# Add a legend
ax.legend()

# Display the chart
plt.show()
```

```python
grouped = data.groupby(['company', 'month'])['Adj close'].mean().unstack()

ax = grouped.plot(kind='bar', figsize=(10, 6))

ax.set_title('Average Stock Prices by Company and Month')
ax.set_xlabel('Month')
ax.set_ylabel('Average Adjusted Close Price')

ax.legend()

plt.show()
```
